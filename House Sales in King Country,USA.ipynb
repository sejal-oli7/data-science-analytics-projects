```python
#Hypothesis Formulation and Statistical Testing
#Research Question

#Does living area significantly affect house price?

#Null Hypothesis (H₀)

#There is no significant relationship between living area and house price.

#Alternative Hypothesis (H₁)

#There is a significant relationship between living area and house price.
```

```python
# Pearson correlation is used because both variables are numerical
corr_coef, p_value = stats.pearsonr(
    df['sqft_living'],
    df['price']
)

print("Correlation:", corr_coef)
print("P-value:", p_value)
```

```python
#Decision Rule
#Interpertation
# p-value <0.05 - significant relationship exists
#p-value > 0.05 → no significant relationship.
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
The dataset was cleaned by handling missing values and removing duplicate records.
House prices are positively skewed with several high-value outliers.
Most houses contain 3–4 bedrooms.
Living area shows a strong positive relationship with house price.
Pearson correlation testing produced a p-value below 0.05, indicating a statistically significant relationship between living area and price.
Larger houses generally sell for higher prices.
```
